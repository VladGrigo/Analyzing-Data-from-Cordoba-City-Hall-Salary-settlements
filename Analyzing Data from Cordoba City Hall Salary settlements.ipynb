## Calculate in which years the largest number of current staff were hired



In this section we will analyze in which year the largest number of employees who are working in the municipality entered. The main task to extract this information is to generate a column containing only the year in which the current employees started working. 

```python
dfp_ing= dfp.copy()
dfp_ing["Ingreso"] = dfp_ing["Ingreso"].astype(str) 
dfp_ing.head()
```

```python
dfp_ing[['Date','Hour']] = dfp_ing["Ingreso"].str.split(expand=True) 
dfp_ing.head()
```

From a survey of the database, we see that there are certain rows within the column "years" that do not contain the year of income but a value that does not resemble any date. We will exclude those rows from the analysis.

```python
dfp_ing_year= dfp_ing.copy()
dfp_ing_year[['Year','Month','Day']] = dfp_ing["Date"].str.split('-', expand=True)
valid_years= (dfp_ing_year['Year'].str.len()==4)
dfp_ing_year=dfp_ing_year[valid_years]
dfp_ing_year.head()
```

```python
dfp_ing_year["Year"]=dfp_ing_year["Year"].astype(int)
dfp_ing_year.groupby("Year").size().plot.line(title="Evolution of the number of employees hired per year")
```

```python
dfp_ing_year["Year"].value_counts().head()
```

By doing a crude analysis of the graph, it can be seen that most of the income of current employees occurred between 2000 and 2010. This conclusion must be taken with care since it only takes into account the workers who are currently employed and not them. who have already retired or left the municipality for various reasons.

## Calculate the number of women and men within the municipality for the generation of gender statistics.

For the purposes of gender analysis, this section will seek to calculate the number of men and women currently working in the municipality. To do this, we will use a job identification method known in Argentina as CUIT. 

The first two numbers of this id correspond to the sex of the corresponding person

- The number 20 corresponds to Male
- The number 27 corresponds to Women

There is also the number 23, 24, 25 and 26 assigned to some people so as not to generate inconsistencies in the creation of the CUIT. Since we cannot know whether they are female or male, we will exclude them from the analysis.

```python
dfp_gender= dfp.copy()
dfp_gender[['Gender','DNI','Last Number']] = dfp_gender["Doc"].str.split('-', expand=True)
dfp_gender.head()
```

```python
dfp_gender["Gender"]=dfp_gender["Gender"].astype(int)
valid_gender= ((dfp_gender['Gender']== 20) | (dfp_gender['Gender']== 27))
dfp_gender=dfp_gender[valid_gender]
dfp_gender.head()
```

```python
dfp_gender.groupby("Gender").size().plot.bar(title="Number of men and women currently working in the municipality")
```

```python
dfp_gender.loc[(dfp_gender['Gender'] == 20),'Gender']='Male'
dfp_gender.loc[(dfp_gender['Gender'] == 27),'Gender']='Female'
dfp_gender.head()
```

```python
dfp_gender.groupby("Gender").size().plot.bar(title="Number of men and women currently working in the municipality")
```

```python
dfp_gender["Gender"].value_counts()
```

## Find the items that are most frequently paid except for net and gross salaries

```python
freq= df["Concepto"].value_counts()
```

```python
freq.head()
```

As we see, the most common concepts paid are made up of the sum of the basic and additional amounts. However, what we are looking for are the most frequent additional liquidated items. Therefore, we need to filter these concepts

First, we filter those concepts that are less than zero

```python
df_freq = df.loc[df['Monto'] > 0]
```

```python
df_freq.info()
```

```python
freq= df_freq["Concepto"].value_counts()
freq.head()
```

```python
exc_conc=["Neto", "MSR", "Básico", "Asignación Basica"]
df_freq = df.loc[(df['Monto'] > 0) & (~df['Concepto'].isin(exc_conc))]
```

```python
df_freq.head()
```

```python
freq= df_freq["Concepto"].value_counts()
freq.head()
```

Excluding the concepts with amounts less than zero and those that are aggregated sums of multiple components, we can conclude that the additional payment for seniority is the most frequent concept among the personnel of the municipality



It would also be important to calculate the most common discounted concept of employee salary. For this we will only need the concepts whose amounts are less than zero

```python
df_freq_dis = df.loc[(df['Monto'] < 0)]
df_freq_dis.head()
```

```python
freq_dis= df_freq_dis["Concepto"].value_counts()
freq_dis.head()
```

Seeing the previous table, social work and retirement contributions are the most discounted concepts from the salary of employees, followed by life insurance, union dues and accident insurance
